```python
import numpy as np
import matplotlib.pyplot as plt

# Example grid
grid = np.array([
    [1, 0, 1, 0],
    [0, 1, 0, 1],
    [1, 1, 0, 0],
    [0, 0, 1, 1]
])

# Grid dimensions
rows, cols = grid.shape

# Plot setup
fig, ax = plt.subplots()
colors = {0: 'red', 1: 'blue'}  # Colors for 0 and 1 in the grid

# Scaling and displacement factors
x_offset = 0.5
y_scale = np.sqrt(3)/2

# Generate coordinates and plot
for row in range(rows):
    for col in range(cols):
        x = col + (x_offset if row % 2 == 0 else 0)
        y = row * y_scale
        color = colors[grid[row, col]]
        ax.plot(x, y, 'o', color=color, markersize=10)  # Adjust markersize as needed

# Set equal scaling and grid visibility
ax.set_aspect('equal')
# plt.grid(True)
plt.show()
```
